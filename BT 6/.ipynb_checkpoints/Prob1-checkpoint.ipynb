Khảo sát tốc độ biến thiên của tín hiệu sin rời rạc $x\left [ n \right ]=A cos\left ( 2\pi fn+\varphi  \right )$

Khi cho giá trị $f$ thay đổi sao cho $\omega \in \left [ -2\pi ,2\pi  \right ]$

```python
# nhập thư viện numpy và matplotlib
import numpy as np
import matplotlib.pyplot as plt


# khởi tạo hàm f
def f(omega, n):
    return np.cos(np.pi*omega*n)
```

```python
# khởi tạo các giá trị omega
omega = [i/10 for i in range(10, 21, 1)]


sampled_n = []

for i in range(50):
    sampled_n.append(i)

res = []
for om in omega:
    _ = [f(om, i) for i in sampled_n]
    res.append(_)
    
fig, axs = plt.subplots(10, 1)


axs[0].stem(sampled_n, res[0])

axs[1].stem(sampled_n, res[1])
# axs[0, 0].set_title('Hàm Y')

axs[2].stem(sampled_n, res[2])

axs[3].stem(sampled_n, res[3])

axs[4].stem(sampled_n, res[4])

axs[5].stem(sampled_n, res[5])

axs[6].stem(sampled_n, res[6])

axs[7].stem(sampled_n, res[7])

axs[8].stem(sampled_n, res[8])

axs[9].stem(sampled_n, res[9])

# axs[10].stem(sampled_n, res[10])

fig.set_size_inches(12,6)


plt.show()
```
