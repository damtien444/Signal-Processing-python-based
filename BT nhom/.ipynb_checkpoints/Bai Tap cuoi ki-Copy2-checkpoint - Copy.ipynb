```python
def find_baseline(data, sampling_rate, OCTAVE):
    i=1
    get_data_slice = data[int((0.5*i-0.5)*WINDOW_SIZE) : int((0.5*i+0.5)*WINDOW_SIZE)]

    fourier_transform = np.fft.rfft(get_data_slice*window, 2**13)

    abs_fourier_transform = np.abs(fourier_transform)
    power_spectrum = np.square(abs_fourier_transform)

    down = []
    for i in range(1, OCTAVE+1):
        down.append(power_spectrum[::i])
    #     down1 = power_spectrum[::1]

    res = []
    for i in range(len(down[-1])):
        mul = 1
        for j in range(0, OCTAVE):
            mul *= down[j][i]
        res.append(mul)

    maxv = max(res)
    return maxv
```

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.io import wavfile

%matplotlib tk

# Detech f0 using the method of Spectrum Harmonic

# Window is the data frame which we're extract from the original file
WINDOW_SIZE = 320

# Innitialize input and constant

# List of final frequency
f0 = []

SPAN = 0
# Number of Harmonic would be include in Calculation
OCTAVE = 2

# Read data from 
sampling_rate, data = wavfile.read('studio_female.wav')

# Innitiallize hanning window with the same size ofthe window
window = np.hanning(WINDOW_SIZE)

# Calculation of basic baseline to filter noise later on
base = find_baseline(data, sampling_rate, OCTAVE)

# Garbage list variable
_ = []

# Iterate through the data with windows that overlaps 50% of each other
for i in range(1, int(len(data)/(WINDOW_SIZE))*2-1):

    # Slice data with the windowsize that overlaps 50%
    get_data_slice = data[int((0.5*i-0.5)*WINDOW_SIZE) : int((0.5*i+0.5)*WINDOW_SIZE)]

    # Generate the fourier power and its corresponding frequency
    fourier_transform = np.fft.rfft(get_data_slice*window, 2**13)
    frequency = np.linspace(0, sampling_rate/2, num=len(fourier_transform))
    
    # Filter the value of too high or too low frequency in the fourier result
    pas = []
    for i in range(len(frequency)):
        if frequency[i]<50 or frequency[i]>5000:
            pas.append(0)
        else:
            pas.append(1)
    for i in range(len(frequency)):
        fourier_transform[i] *= pas[i]

    # Compute Power
    abs_fourier_transform = np.abs(fourier_transform)
    power_spectrum = np.square(abs_fourier_transform)

    # Downsampling with the numnber of octave, then save to the list down
    down = []
    for i in range(1, OCTAVE+1):
        down.append(power_spectrum[::i])
    #     down1 = power_spectrum[::1]

    # Multiply vertically thru the downsample version to create the greatest peak
    res = []
    for i in range(len(down[-1])):
        mul = 1
        for j in range(0, OCTAVE):
            mul *= down[j][i]
        res.append(mul)
    
    # Find the index of the max peak of the multiply downsample version
    maxv = max(res)
    
    if maxv < base:
        indv = 0
    
    indv = res.index(maxv)
    
    if frequency[indv] < 90 or frequency[indv] > 350:
        _.append(0)
        continue
    else:
        _.append(np.mean([frequency[i] for i in range(indv-SPAN, indv+SPAN+1, 1)]))
        
        print([frequency[i] for i in range(indv-SPAN, indv+SPAN+1, 1)])

print(_)
time = []
for i in range(len(_)):
    time.append(int(i*WINDOW_SIZE-WINDOW_SIZE/2))

plt.plot(time, _, 'o')
plt.plot(0, 0, 'o')
```
