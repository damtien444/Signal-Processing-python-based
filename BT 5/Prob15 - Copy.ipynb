## Section a:

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('lena.jpg')
imgplot = plt.imshow(img, cmap="gray")
img
```

## Section b:

As the requirements, this is equivalent to the function medfilt2d in scipy.signal library. This function will apply a median filter to the input array using a local window-size given by kernel_size (must be odd). The array is zero-padded automatically.

```python
from scipy.signal import medfilt2d

img33 = medfilt2d(img)
imgplot = plt.imshow(img33, cmap="gray")
```

## Section c:

```python
from scipy.signal import medfilt2d

img55 = medfilt2d(img, 5)
imgplot = plt.imshow(img55, cmap="gray")
```

## Comment:

The image is blurred by both medium filters, the second filter make the image more blur.

```python
fig, axs = plt.subplots(3)

fig.set_size_inches(12,12)


axs[0].imshow(img, cmap="gray")
axs[1].imshow(img33, cmap="gray")
axs[2].imshow(img55, cmap="gray")

```
