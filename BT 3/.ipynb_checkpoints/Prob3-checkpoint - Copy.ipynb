```python
from os.path import dirname, join as pjoin
from scipy.io import wavfile
from scipy.io.wavfile import write
import scipy.io
import scipy.signal as sps
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Audio
```

```python
samplerate, data = wavfile.read('violin-origional.wav')
# print(f"number of channel = {data.shape[1]}")
```

```python
length = data.shape[0] / samplerate
# Trong Data.shape là một tuple có 2 dữ liệu
# [1] là số channel của data

print(f"Length in second = {length}s")
print(f"Sample rate = {samplerate}")
print(f"Number of samples = {data.shape[0]}")
```

```python
time = np.linspace(0., length, data.shape[0])
# Sinh ra danh sách timestamp cho sample dữ liệu cường độ 

plt.plot(time, data, label="Violin")
# Chỉ lấy dữ liệu của kênh bên trái

plt.legend()
# Hiện chú thích

plt.xlabel("Thời gian [s]")
plt.ylabel("Cường độ")
```

```python
Audio(data, rate=samplerate)
```

```python
Audio(data, rate=samplerate/2)
```

```python
Audio(data, rate=samplerate*2)
```

```python
fdiv2_rate =  samplerate/2
fdiv2_number_of_sample = round(len(data) / 2)

newdata = sps.resample(data, fdiv2_number_of_sample)
newtime = np.linspace(0., length, newdata.shape[0])

plt.plot(newtime, newdata, label="Violin")
plt.legend()
# Hiện chú thích

plt.xlabel("Thời gian [s]")
plt.ylabel("Cường độ")

Audio(newdata, rate=fdiv2_rate)
```

```python
fmul2_rate =  samplerate*2
fmul2_number_of_sample = round(len(data) * 2)

newdata = sps.resample(data, fmul2_number_of_sample)
newtime = np.linspace(0., length, newdata.shape[0])

plt.plot(newtime, newdata, label="Violin")
plt.legend()
# Hiện chú thích

plt.xlabel("Thời gian [s]")
plt.ylabel("Cường độ")

Audio(newdata, rate = fmul2_rate)
```
