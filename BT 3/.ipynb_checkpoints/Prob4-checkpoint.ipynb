```python
from thinkdsp import*
```

```python
sin_sig = SinSignal(freq=440, amp=1.0, offset=0)
sin_sig.plot()
wave = sin_sig.make_wave(duration=1, framerate=10*440)
wave.apodize()
wave.make_audio()
```

```python
duration = sin_sig.period*3
segment = sin_sig.make_wave(duration, framerate=30*440)
segment.plot()
decorate(xlabel='Time (s)', ylabel="Amptitude")
wave = sin_sig.make_wave(duration=1, framerate=30*440)
wave.apodize()
wave.make_audio()
```

```python
duration = sin_sig.period*3
segment = sin_sig.make_wave(duration, framerate=1.5*10*440)
segment.plot()
decorate(xlabel='Time (s)', ylabel="Amptitude")
wave = sin_sig.make_wave(duration=1, framerate=1.5*10*440)
wave.apodize()
wave.make_audio()
```
