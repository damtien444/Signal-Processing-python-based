```python
import matplotlib.pyplot as plt
from numpy import heaviside

def u(x):
    if x>= 0:
        return 1
    else:
        return 0
    
def y(n):
    return (2*u(n-2)-2*u(n-7)-2*u(-n)+2*u(4-n))
    
def cau_a(n):
    return 2-3*y(n)

def cau_b(n):
    return 3*y(n-2)

def cau_c(n):
    return 2-2*y(-2+n)
```

```python
default = []
for i in range(-4,11):
    default.append(y(i))

a = []
for i in range(-4,11):
    a.append(cau_a(i))

b = []
for i in range(-4,11):
    b.append(cau_b(i))
    
c = []
for i in range(-4,11):
    c.append(cau_c(i))
    
print(default)
```

```python
plt.xlabel("thời gian(n)")
plt.ylabel("Amplitude")
plt.title("Bài 2, câu a")
plt.stem(range(-4,11), a)
```

```python
plt.xlabel("thời gian(n)")
plt.ylabel("Amplitude")
plt.title("Bài 2, câu b")
plt.stem(range(-4,11), b)
```

```python
plt.xlabel("thời gian(n)")
plt.ylabel("Amplitude")
plt.title("Bài 2, câu c")
plt.stem(range(-4,11), c)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].stem(range(-4,11), default)
axs[0, 0].set_title('Hàm Y')
axs[0, 1].stem(range(-4,11), a, 'tab:orange')
axs[0, 1].set_title('Câu A')
axs[1, 0].stem(range(-4,11), b, 'tab:green')
axs[1, 0].set_title('Câu B')
axs[1, 1].stem(range(-4,11), c, 'tab:red')
axs[1, 1].set_title('Câu C')

for ax in axs.flat:
    ax.set(xlabel='Thời gian', ylabel='Cường độ')

# Hide x labels and tick labels for top plots and y ticks for right plots.
# for ax in axs.flat:
#     ax.label_outer()
fig.tight_layout()
```
