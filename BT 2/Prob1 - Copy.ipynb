```python
import matplotlib.pyplot as plt

def u(x, shift=0):
    if (x>=(0 + shift)):
        return 1
    else:
        return 0

def x(n):
    de = [-2, 4, -2, 2, -1]
    if n>= -1 and n <=3:
        return de[n+1]
    else:
        return 0
    
def cau_a(n):
    return 3*x(-n-1)

def cau_b(n):
    return x(2*n)-1

def cau_c(n):
    return -x(n)+2
```

```python
default = []
for i in range(-10,11):
    default.append(x(i))
```

```python
a = []
for i in range(-10,11):
    a.append(cau_a(i))
```

```python
b = []
for i in range(-10,11):
    b.append(cau_b(i)) 
```

```python
c = []
for i in range(-10,11):
    c.append(cau_c(i))
```

```python
plt.xlabel("Time(n)")
plt.ylabel("Amtitude")
plt.title('Câu a')
plt.stem(range(-10,11), a)
```

```python
plt.xlabel("Time(n)")
plt.ylabel("Amtitude")
plt.title('Câu b')
plt.stem(range(-10,11), b)
```

```python
plt.xlabel("Time(n)")
plt.ylabel("Amtitude")
plt.title('Câu c')
plt.stem(range(-10,11), c)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].stem(range(-10,11), default)
axs[0, 0].set_title('Hàm Y')
axs[0, 1].stem(range(-10,11), a, 'tab:orange')
axs[0, 1].set_title('Câu A')
axs[1, 0].stem(range(-10,11), b, 'tab:green')
axs[1, 0].set_title('Câu B')
axs[1, 1].stem(range(-10,11), c, 'tab:red')
axs[1, 1].set_title('Câu C')

for ax in axs.flat:
    ax.set(xlabel='Thời gian', ylabel='Cường độ')

# Hide x labels and tick labels for top plots and y ticks for right plots.
# for ax in axs.flat:
#     ax.label_outer()
fig.tight_layout()
```
