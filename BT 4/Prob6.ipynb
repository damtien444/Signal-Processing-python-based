# Problem 6

The problem use the sound file "handel" available in $Matlab$. This sound is sampled at $F_{s}=8192$ samples per second using 8-bits per sample.

a) Load the sound waveform “handel” in an array $x$ and listen to it using the sound function at the full sampling rate.

b) Select every other sample in x which reduces the sampling rate by a factor of two. Now listen to the new sound array using the sound function at half the sampling rate.

c) Select every fourth sample in x which reduces the sampling rate by a factor of four. Listen to the resulting sound array using the sound function at quarter the sampling rate.

d) Save the generated sound in part (c) using the wavwrite function

# Section a:

```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Audio

# read the handel to an array data with samplerate in the variable of "samplerate"
samplerate, data = wavfile.read('handel.wav')

# playback it with audio function at the normal sampling rate
Audio(data, rate=samplerate)
```

# Section b:

```python
from scipy.signal import decimate
import scipy.signal as sps

# downsampling to a factor of two, the function required 2 paras, 
# 1st: The signal to be downsampled, as an N-dimensional array.
# 2nd: The downsampling factor

data_down = decimate(data, 2)

# playback with the framerate haft as the normal sampling rate.
Audio(data_down,rate =  int(samplerate/2))
```

### Comment: 

The signal that has lower pitch everytime it has been downsampled of the factor of two. 

# Section c:

```python
from scipy.signal import decimate

# downsampling to a factor of two, the function required 2 paras, 
# 1st: The signal to be downsampled, as an N-dimensional array.
# 2nd: The downsampling factor
data_down = decimate(data, 4)

# playback with the framerate haft as the normal sampling rate.
Audio(data_down, rate = int(samplerate/4))
```

*Kindly note that: the sampling rate is very unusual so that the default player of browser **can't play** it, so we have to download it to play with the system player.*

# Section d:

```python
wavfile.write("out.wav", rate=int(samplerate/4), data=data_down)
```
