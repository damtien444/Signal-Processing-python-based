# Problem 7

Compute and plot the response of the following systems:

a) $y\left [ n \right ] = \frac{n}{n+1}y\left [ n -1\right ] +x\left [ n \right ]$, $y\left [ -1 \right ]=0$

b) $y\left [ n \right ] = 0.9y\left [ n -1\right ] +x\left [ n \right ]$, $y\left [ -1 \right ]=0$

to the inputs $y\left [ n \right ] = \delta \left [ n \right ]$ and $y\left [ n \right ] = \delta \left [ n - 5 \right ]$, for $-20\leq n \leq 20$ and comment upon the obtained results.

## Prepare functions and libraries

```python
# Plot library
import matplotlib.pyplot as plt

# Function unit impulse x for both section, using the parameter mode to decide which section is it running in.
def x(n, mode=0):
    if mode == 0:
        if n == 0:
            return 1
        else:
            return 0
    else:
        if n == 5:
            return 1
        else:
            return 0
    
# Function y as the section a. required
def y_a(n, mode=0):
    if n == -1:
        return 0
    else:
        return (n/(n+1))*y_a(n-1, mode) + x(n, mode)
    
# Function y as the section b. required
def y_b(n, mode=0):
    if n == -1:
        return 0
    else:
        return (0.9)*y_b(n-1, mode) + x(n, mode)
```

## Section a:

```python
# Create sample for normal impulse signal:
a0 = []
for i in range(0,21):
    a0.append(y_a(i, mode = 0))
    
# Create sample for shifted 5 step impulse signal:
a1 = []
for i in range(0,21):
    a1.append(y_a(i, mode = 1))
    
fig = plt.figure()
fig.set_size_inches(12,6)
plt.subplots_adjust(hspace=1)
plt.stem(range(0, 21, 1), a0)
plt.title('x[n]=delta[n]')
plt.xlabel('Time [n]')
plt.ylabel('Value')

fig = plt.figure()
fig.set_size_inches(12,6)
plt.subplots_adjust(hspace=2)
plt.stem(range(0, 21, 1), a1)
plt.title('x[n]=delta[n-5]')
plt.xlabel('Time [n]')
plt.ylabel('Value')
```

### Comment: 

The unit impulse initializes the whole signal, the patern is different because of the propertises of the original signal, that's the second graph has lower slope angle, the first graph has higher slope angle therefore the signal's amptitude drop quicklier.

Beside, the impulse signal has effect at different points, because of the original setting at 5.

## Section b:

```python
b0 = []
for i in range(0,21):
    b0.append(y_b(i, mode = 0))
    
b1 = []
for i in range(0,21):
    b1.append(y_b(i, mode = 1))
    
fig = plt.figure()
fig.set_size_inches(12,6)
plt.subplots_adjust(hspace=1)
plt.stem(range(0, 21, 1), b0)
plt.title('x[n]=delta[n]')
plt.xlabel('Time [n]')
plt.ylabel('Value')

fig = plt.figure()
fig.set_size_inches(12,6)
plt.subplots_adjust(hspace=2)
plt.stem(range(0, 21, 1), b1)
plt.title('x[n]=delta[n-5]')
plt.xlabel('Time [n]')
plt.ylabel('Value')
```
